```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Condições iniciais

```python
Lx = 1  # comprimento da placa em cm
Ly = 1  # largura da placa em cm
T = 0.5  # tempo total em segundos
nx = 50
ny = 50
nt = 100

dx = Lx/(nx-1)  # discretização em x em cm
dy = Ly/(ny-1)  # discretização em y em cm
dt = T/nt
c = 1

x = np.linspace(0, Lx, nx)
y = np.linspace(0, Ly, ny)

u = np.zeros((nx, ny, nt+1))

# Condições iniciais e de contorno
u[:, :, 0] = np.sin(np.pi*x) * np.sin(np.pi*y)
```

## Equação da Onda

```python
for n in range(1, nt):
    for i in range(1, nx - 1):
        for j in range(1, ny - 1):
          u[i,j, n+1] = 2 * u[i,j, n] - u[i,j, n-1] + (c * dt/dx)**2 * (u[i+1,j, n] - 2 * u[i,j, n] + u[i-1,j, n]) + (c * dt/dy)**2 * (u[i,j+1, n] - 2 * u[i,j, n] + u[i,j-1, n])
```

## Gráfico de calor 3D

```python
from matplotlib import cm

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111, projection = '3d')


X, Y = np.meshgrid(x, y)
Z = u[:, :, 0]

for n in range(0, n+1, int(nt/10)):
    ax.plot_surface(X, Y, u[:, :, n], cmap=cm.jet, rstride=1, cstride=1,
                       linewidth=0, antialiased=False)


fig.colorbar(ax.plot_surface(X, Y, u[:, :, n], cmap=cm.jet, rstride=1, cstride=1,
                       linewidth=0, antialiased=False), shrink=0.5, aspect=10)


ax.set_title('3D Heatmap in 100/10 points')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('u-axis')
plt.show()
```
