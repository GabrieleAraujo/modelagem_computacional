```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def edos(t, y):
    S, E1, E2, I, Ia, R = y
    N = S + E1 + E2 + I + Ia + R
    
    dSdt = -beta * S * (I + Ia) / N - r_ia * beta * S * Ia / N - r_ea * beta * S * E2 / N
    dE1dt = beta * S * (I + Ia) / N + r_ia * beta * S * Ia / N + r_ea * beta * S * E2 / N - lambda1 * E1
    dE2dt = lambda1 * (1 - pa) * E1 - lambda2 * E2
    dIdt = lambda2 * E2 - u * I
    dIadt = lambda1 * pa * E1 - u * Ia
    dRdt = u * (I + Ia)
    
    return [dSdt, dE1dt, dE2dt, dIdt, dIadt, dRdt]
```

```python
def runge_kutta4(f, t0, y0, h, n):
    t = np.zeros(n)
    y = np.zeros((n, len(y0)))

    t[0] = t0
    y[0] = y0

    for i in range(n - 1):
        k1 = h * np.array(f(t[i], y[i]))
        k2 = h * np.array(f(t[i] + h / 2, y[i] + k1 / 2))
        k3 = h * np.array(f(t[i] + h / 2, y[i] + k2 / 2))
        k4 = h * np.array(f(t[i] + h, y[i] + k3))
        
        y[i + 1] = y[i] + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        t[i + 1] = t[i] + h

    return t, y
```

### Par√¢metros e Condi√ß√µes Iniciais

```python
# Par√¢metros
beta = 0.5
r_ia = 0.1
r_ea = 1.15
lambda1 = 0.33
lambda2 = 0.5
pa = 0.4
u = 0.2

#COndi√ß√µes iniciais
S0 = 5e6
E1_0 = 0.0
E2_0 = 0.0
I0 = 1.0
Ia0 = 0.0
R0 = 0.0

# Total da popula√ß√£o (N)
N = S0 + E1_0 + E2_0 + I0 + Ia0 + R0

# Vetor de condi√ß√µes iniciais
y0 = [S0, E1_0, E2_0, I0, Ia0, R0]

# Tempo de simula√ß√£o
t_start = 0
t_end = 120
dt = 1.0
n = int((t_end - t_start) / dt) + 1

# t = np.arange(0, 100, 0.1)
h = dt
# n = len(t)
```

Popula√ß√£o (N), a qual refere-se √† rela√ß√£o ou fra√ß√£o de uma determinada parte de uma popula√ß√£o em rela√ß√£o ao total, ou seja, pode variar de 0 a 1. De forma que propor√ß√£o de indiv√≠duos infectados(I)/popula√ß√£o total.
Conforme https://www.mdpi.com/2227-7390/9/17/2076, o n√∫mero da popula√ß√£o √© dado por M = ùëÜ(0)+ùê∏(0)+ùêº(0)+ùëÖ(0).

```python
t, y = runge_kutta4(edos, t_start, y0, h, n)
```

```python

plt.figure(figsize=(10, 6))
plt.plot(t, y[:, 0], label=" Suscept√≠veis (S)")
plt.plot(t, y[:, 1], label="Expostos 1 (E1)")
plt.plot(t, y[:, 2], label="Expostos 2 (E2)")
plt.plot(t, y[:, 3], label="Infectados (I)")
plt.plot(t, y[:, 4], label="Infectados assintom√°ticos (Ia)")
plt.plot(t, y[:, 5], label="Removidos (R)")
plt.legend(title = 'Categorias')
plt.xlabel("Tempo")
plt.ylabel("Popula√ß√£o")
plt.title("Modelo SEIR por meio do m√©todo Runge-Kutta de quarta ordem")
plt.grid(True)
plt.show()
```
