```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Experimento I 

```python
# ParÂmetros
L = 20
T = 30
J = 20
N = 60
sigma = 0.5

deltax = L / J
deltat = T / N
x = np.arange(0, L + deltax, deltax)
t = np.arange(0, T + deltat, deltat)

u = np.zeros((J + 1, N + 1))

#constantes
theta = 1250
sigma = deltat / deltax**2

#condição inicial
u[0, 0] = 0
u[J, 0] = 0
u[:, 0] = (-(4 * theta / L**2) * x * (x - L))[:J+1]

for n in range(1, N + 1):
    for j in range(1, J):
        u[j, n] = (1 - 2 * sigma) * u[j, n - 1] + sigma * (u[j + 1, n - 1] + u[j - 1, n - 1])
```

```python
fig1 = plt.figure()
ax = fig1.add_subplot(111, projection='3d')
surf = ax.plot_surface(np.tile(t, (J + 1, 1)), np.transpose(np.tile(x, (N + 1, 1))), u, cmap='viridis')
ax.set_xlabel('Tempo')
ax.set_ylabel('Espaço')
ax.set_zlabel('Temperatura')
ax.set_title('(a)')
#colorbar
fig1.colorbar(surf, ax=ax, pad=0.1, aspect=10)

fig2, ax2 = plt.subplots()
ax2.plot(x, u[:, N], linewidth=2.0)
ax2.grid(True)
ax2.set_xlabel('Espaço')
ax2.set_ylabel('Temperatura')
ax2.set_title('(b)')

plt.show()
```
