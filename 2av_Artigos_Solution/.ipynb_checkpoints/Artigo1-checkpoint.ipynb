# Paper - Fuzzy system model for gene expression

- https://doi.org/10.1016/j.ejmhg.2018.06.002

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Funções

```python
#Funções EDO:
def f_w(t, w, x, p, k1, k2):
    return -k1 * w

def f_x(t, w, x, p, k1, k2):
    return k1 * w - k2 * x

def f_p(t, w, x, p, k2):
    return k2 * x
```

```python
#Analíticas
def anat_w(t, w0, k1):
    return w0*np.exp(-k1*t)

def anat_x(t, w0, x0, k1, k2):
    return (w0*k1 / (k2-k1))*(np.exp(-k1*t) - np.exp(-k2*t)) + x0*np.exp(-k2*t)

def anat_p(t, w0, x0, p0, k1, k2):
    return w0*(1 + ((k1*np.exp(-k2*t) - np.exp(-k2)*np.exp(-k1*t)) / (k2-k1))) + x0*(1-np.exp(-k2*t))+p0
```

## Condições Iniciais

```python
#Condições iniciais
t = np.arange(0, 25, 0.2)  #25s
h = 0.1  
n = len(t)
crisp_initial = 25  # Condições iniciais de DNA, mRNA e proteína
k1 = 0.42  #taxa de transcrição (microgramas/s)
k2 = 2.59  #taxa de trasnlação (microgramas/s)
```

## Método de Euler

```python
def euler(t, w0, x0, p0, h, n, k1, k2):
    w = np.zeros(n)
    x = np.zeros(n)
    p = np.zeros(n)
    w[0] = w0
    x[0] = x0
    p[0] = p0
    for i in range(n - 1):
        w[i + 1] = w[i] + f_w(t[i], w[i], x[i], p[i], k1, k2) * h
        x[i + 1] = x[i] + f_x(t[i], w[i], x[i], p[i], k1, k2) * h
        p[i + 1] = p[i] + f_p(t[i], w[i], x[i], p[i], k2) * h

    return w, x, p
```

```python
w, x, p = euler(t, crisp_initial, crisp_initial, crisp_initial, h, n, k1, k2)
```

### Gráfico

```python
def plot_euler(t, w, x, p):
    plt.figure(figsize=(10, 6))
    plt.scatter(t, w, label='Concentração de DNA (w) - Euler')
    plt.scatter(t, x, label='Concentração de mRNA (x) - Euler')
    plt.scatter(t, p, label='Concentração de proteína (p) - Euler')
    plt.xlabel('Time')
    plt.ylabel('Concentrações de DNA, mRNA e Proteína')
    plt.legend()
    plt.title('Euler - The graphical representation of α-cut of the solution set X')
    plt.grid(True)
    plt.show()
plot_euler(t, w, x, p)
```

# Runge-Kutta

```python
def runge_kutta(t, w0, x0, p0, h, n, k1, k2):
    w = np.zeros(n)
    x = np.zeros(n)
    p = np.zeros(n)
    w[0] = w0
    x[0] = x0
    p[0] = p0

    for i in range(n - 1):
#         k1 = f(x[i], y[i])
#         k2 = f(x[i] + h/2, y[i] + h/2 * k1)
#         k3 = f(x[i] + h/2, y[i] + h/2 * k2)
#         k4 = f(x[i] + h, y[i] + h * k3)
#         y[i + 1] = y[i] + h/6 * (k1 + 2*k2 + 2*k3 + k4)
        
        k1w = h * f_w(t[i], w[i], x[i], p[i], k1, k2)
        k1x = h * f_x(t[i], w[i], x[i], p[i], k1, k2)
        k1p = h * f_p(t[i], w[i], x[i], p[i], k2)

        k2w = h * f_w(t[i] + h/2, w[i] + k1w/2, x[i] + k1x/2, p[i] + k1p/2, k1, k2)
        k2x = h * f_x(t[i] + h/2, w[i] + k1w/2, x[i] + k1x/2, p[i] + k1p/2, k1, k2)
        k2p = h * f_p(t[i] + h/2, w[i] + k1w/2, x[i] + k1x/2, p[i] + k1p/2, k2)

        k3w = h * f_w(t[i] + h/2, w[i] + k2w/2, x[i] + k2x/2, p[i] + k2p/2, k1, k2)
        k3x = h * f_x(t[i] + h/2, w[i] + k2w/2, x[i] + k2x/2, p[i] + k2p/2, k1, k2)
        k3p = h * f_p(t[i] + h/2, w[i] + k2w/2, x[i] + k2x/2, p[i] + k2p/2, k2)

        k4w = h * f_w(t[i] + h, w[i] + k3w, x[i] + k3x, p[i] + k3p, k1, k2)
        k4x = h * f_x(t[i] + h, w[i] + k3w, x[i] + k3x, p[i] + k3p, k1, k2)
        k4p = h * f_p(t[i] + h, w[i] + k3w, x[i] + k3x, p[i] + k3p, k2)

        w[i + 1] = w[i] + (k1w + 2*k2w + 2*k3w + k4w) / 6
        x[i + 1] = x[i] + (k1x + 2*k2x + 2*k3x + k4x) / 6
        p[i + 1] = p[i] + (k1p + 2*k2p + 2*k3p + k4p) / 6

    return w, x, p
```

```python
w_runge_kutta, x_runge_kutta, p_runge_kutta = runge_kutta(t, crisp_initial, crisp_initial, crisp_initial, h, n, k1, k2)
```

```python
def plot_runge_kutta_results(t, w_runge_kutta, x_runge_kutta, p_runge_kutta):
    plt.figure(figsize=(10, 6))
    plt.scatter(t, w_runge_kutta, label='Concentração de DNA (w) - Runge-Kutta')
    plt.scatter(t, x_runge_kutta, label='Concentração de mRNA (x) - Runge-Kutta')
    plt.scatter(t, p_runge_kutta, label='Concentração de proteína (p) - Runge-Kutta')
    plt.xlabel('Time')
    plt.ylabel('Concentrações de DNA, mRNA e Proteína')
    plt.legend()
    plt.title('Runge-Kutta - The graphical representation of α-cut of the solution set X')
    plt.grid(True)
    plt.show()
plot_runge_kutta_results(t, w_runge_kutta, x_runge_kutta, p_runge_kutta)
```

# Comparação Euler e Runge-Kutta

```python
plt.figure(figsize=(10, 6))
plt.scatter(t, w, label='Concentração de DNA (w) - Euler')
plt.scatter(t, x, label='Concentração de mRNA (x) - Euler')
plt.scatter(t, p, label='Concentração de proteína (p) - Euler')
plt.scatter(t, w_runge_kutta, label='Concentração de DNA (w) - Runge-Kutta')
plt.scatter(t, x_runge_kutta, label='Concentração de mRNA (x) - Runge-Kutta')
plt.scatter(t, p_runge_kutta, label='Concentração de proteína (p) - Runge-Kutta')
plt.xlabel('Time')
plt.ylabel('Concentrações de DNA, mRNA e Proteína')
plt.legend()
plt.title('Euler vs Runge-Kutta')
plt.grid(True)
plt.show()
```

# Comparação dos métodos com a Analítica

```python
# Funções analíticas
w_anat = lambda t: anat_w(t, crisp_initial, k1)
x_anat = lambda t: anat_x(t, crisp_initial, crisp_initial, k1, k2)
p_anat = lambda t: anat_p(t, crisp_initial, crisp_initial, crisp_initial, k1, k2)
```

```python
def plot_results(t, w, x, p, w_runge_kutta, x_runge_kutta, p_runge_kutta, w_anat, x_anat, p_anat):
    plt.figure(figsize=(12, 8))
    plt.plot(t, w, label='DNA (w) - Euler', marker='o')
    plt.plot(t, x, label='mRNA (x) - Euler', marker='o')
    plt.plot(t, p, label='Proteína (p) - Euler', marker='o')
    plt.plot(t, w_runge_kutta, label='DNA (w) - Runge-Kutta', marker='o')
    plt.plot(t, x_runge_kutta, label='mRNA (x) - Runge-Kutta', marker='o')
    plt.plot(t, p_runge_kutta, label='Proteína (p) - Runge-Kutta', marker='o')
    plt.plot(t, w_anat(t), linestyle='--', label='DNA (w) - Analítica', marker='o')
    plt.plot(t, x_anat(t), linestyle='--', label='mRNA (x) - Analítica', marker='o')
    plt.plot(t, p_anat(t), linestyle='--', label='Proteína (p) - Analítica', marker='o')
    plt.xlabel('Time')
    plt.ylabel('Concentrações de DNA, mRNA e Proteína')
    plt.legend()
    plt.title('Métodos - Euler vs Runge Kutta vs Analítico')
    plt.grid(True)
    plt.show()
    
plot_results(t, w, x, p, w_runge_kutta, x_runge_kutta, p_runge_kutta, w_anat, x_anat, p_anat)
```
