```python
import numpy as np
```

```python
#(1-4/3*t)*y
```

```python
def f(x, y): 
    return (1-(4/3)*x)*y
```

```python
def df(x,y):
    return -(4/3)*y
```

```python
def analitica(x):
    return np.exp(x-(2/3)*x**2)
```

```python
def calcular_prox_valor(a, b, N, y0, x):
    y = np.zeros(N)
    h=(b-a)/N
    y[0] = y0
    max_inter = 6
    tolerancia = 0.001
    for i in range(0, N-1):
        erro = 1
        zold = y[i] + h*f(x[i], y[i])
        j = 0
    
        while erro > tolerancia and j <= max_inter:
            F = y[i] + h * f(x[i+1], zold) - zold
            dF = h * df(x[i+1], zold) - 1
            xnew = zold - F/dF
            erro = abs (xnew - zold)
            zold = xnew
            j = j+1
        y[i+1] = xnew
    return y
```

# Condições Iniciais 

```python
#domínios
a  = 0
b  = 1
y0 = 1  #PVI confição inicial
N  = 50   # Número de interações
```

```python
x = np.linspace(a, b, N) #domínio
```

```python
F = calcular_prox_valor(a, b, N, y0, x)
print (f'Solução Euler: {F}')
len(F)
```

```python
analic = analitica(x)
print (f'Solução Analítica: {analic}')
len(analic)
```
