# Adams-Bashforth 3 passos

```python
def bashforth(x, y0, h):
    y = np.zeros(len(x))
    y[0] = y0
    
    K1 = f(x[1], y[1])
    K2 = f(x[0], y[0])
    y[1] = y[0] +  h *(K2)
    y[2] = y[1] +  h *(K1)
    
    for i in np.arange(2, len(x)-1):
        K3 = K2
        K2 = K1
        K1=f(x[i],y[i])
        y[i+1] = y[i] + h*(23* K1 -16* K2 + 5* K3 ) /12
                
    return y    
```

```python
E = bashforth(x, y0, h)
print (f'Solução Euler: {E}')
len(E)
```

## Range-Kutta - Método de Barshforth com 4 passos

```python
def range_kutta(x, y0, h):
    y = np.zeros(len(x))
    y[0] = y0
    
    for i in np.arange(3, len(x)-1):
          
            k1 = f(x[i],y[i])
            k2 = f(x[i] + (h/2), y[i] + (h/2) * k1)
            k3 = f(x[i] + h, y[i] + h * k2)
            k4 = f(x[i] + h, y[i] + h * k3)
            
            y[i+1] = y[i] + (h/6) * (k1 + 2 * k2 + 2 * k3 + k4)
#             x[i+1] = x[i] + h
            print (y[i+1])
                
    return y
```

```python
def rk4(f,x,y,h,n) :
    for _ in range (n):
        k1 = h * f (x, y)
        k2 = h * f ( x + 0.5* h , y + 0.5* k1 )
        k3 = h * f ( x + 0.5* h , y + 0.5* k2 )
        k4 = h * f ( x + h , y + k3 )
        y += (k1 + 2* k2 + 2* k3 + k4 ) / 6
        x += h
    return y
```

```python
def adams(x, y0, h):
    _y = np.zeros(len(x))
    _y[0] = y0
    
    for i in np.arange (0 , 3) :
        _y[i+1] = rk4(f,x[i],_y[i],h,2)

    f_m2 = f(x[0] , y[0])
    f_m1 = f(x[1] , y[1])
    f_0 = f(x[2] , y[2])
    f_1 = f(x[3] , y[3])

    for i in range (3, len(x)- 1) : # Inicia no 3
        f_m3 = f_m2
        f_m2 = f_m1
        f_m1 = f_0
        f_0 = f_1

        _y[i+1] = _y[i]+(h/24) * (55* f_0 - 59* f_m1 + 37* f_m2 - 9* f_m3 )
        f_1 = f (x[i+1] , _y[i+1])

    return _y
```

```python
R = adams(x, y0, h)
print (f'Solução Range Kutta: {R}')
len(R)
```
