# Laplace Bidimensional (Aula 27/10)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Condições iniciais

```python
# Condições iniciais
Lx = 2*np.pi
Ly = 2*np.pi
T = 0.5
c = 1.0  # velocidade da onda
nx = 101  # Número de pontos na direção x
ny = 101
nt = 1000  # Número de passos no tempo
dx = 2*np.pi/(nx - 1)  # Delta x
dy = 2*np.pi/(ny - 1)  # Delta x
dt = T / nt  # Delta t


x = np.linspace(0, np.pi, nx)
y = np.linspace(0, np.pi, ny)
u = np.zeros((nx, ny))
u[:, 0] = np.sin(x)
u[:,-1] = np.sin(y)
```

```python
for n in range(1000):
    for i in range(1, nx - 1):
        for j in range(1, ny - 1):
            u[i, j] = 0.25 * (u[i + 1, j] + u[i - 1, j] + u[i, j + 1] + u[i, j - 1])
```

```python
#Matriz resultante
plt.imshow(u, extent=[0, Lx, 0, Ly], origin='lower', aspect='auto', cmap='viridis')
plt.colorbar()
plt.title('Laplace Bidimensional')
plt.xlabel('Direção x')
plt.ylabel('Direção y')
plt.show()
```

```python
for n in range(0, nt, int(nt / 10)):  # Definindo alguns pontos
    plt.figure(figsize=(6, 6))
    plt.imshow(u[:, :, n], cmap='viridis', extent=[0, Lx, 0, Ly], origin='lower', aspect='auto')
    plt.colorbar()
    plt.title(f"Tempo: {n * dt:.2f}")
    plt.show()
```
