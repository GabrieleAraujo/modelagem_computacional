# MÃ©todo de Adam-Bashfort (Aula 13/10)

```python
import numpy as np
```

```python
#AnalÃ­tica
def analitf(t): 
    return (t+1.0)**2 - np.exp(t)/2.0
```

```python
#Derivada
def f(tau, W): 
    return W - tau**2 + 1.0
```

## MÃ©todo de Adams-Bashforth - 2 passos

```python
def euler_bashforth(x, y0, h):
    y = np.zeros(len(x))
    y[0] = y0

    K1 = f(x[0] , y[0])
    y [1] = y [0] +  h *(K1)

    for i in np.arange(1, len(x)-1):
        K2 = K1
        K1=f(x[i],y[i])
        y[i+1] = y[i] + h *(3* K1 - K2) /2
                
    return y         
```

```python
#domÃ­nios
a  = 0.0
b  = 2.0
y0 = 0.5  #PVI confiÃ§Ã£o inicial
n  = 0   # NÃºmero de interaÃ§Ãµes
h  = 0.001  #intervalo (delta) h=(b-a)/100
#h  = 0.1 
```

```python
x = np.arange(a, b, h) #ou t
len(x)
```

```python
F = euler_bashforth(x, y0, h)
print (f'SoluÃ§Ã£o Euler: {F}')
len(F)
```

```python
analic = analitf(x)
print (f'SoluÃ§Ã£o AnalÃ­tica: {analic}')
len(analic)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(font_scale = 1, font='sans serif')
plt.figure(figsize=(10,6), facecolor = "#FFF")

plt.figure(1)
plt.rcParams.update({'font.size': 10})
plt.plot(x, F, 'r--', linewidth = 1)
plt.plot(x, analic, 'b--', linewidth = 2)
plt.title('Euler')
plt.xlabel('x')
plt.ylabel('y(x)')
#plt.legen(['Euler'])
plt.grid(True)
plt.show()
```

```python
def bashforth(x, y0, h):
    y = np.zeros(len(x))
    y[0] = y0
    
    K1 = f(x[1], y[1])
    K2 = f(x[0], y[0])
    y[1] = y[0] +  h *(K2)
    y[2] = y[1] +  h *(K1)
    
    for i in np.arange(2, len(x)-1):
        K3 = K2
        K2 = K1
        K1=f(x[i],y[i])
        y[i+1] = y[i] + h*(23* K1 -16* K2 + 5* K3 ) /12
                
    return y    
```

```python
def rk4(f,x,y,h,n) :
    for _ in range (n):
        k1 = h * f (x, y)
        k2 = h * f ( x + 0.5* h , y + 0.5* k1 )
        k3 = h * f ( x + 0.5* h , y + 0.5* k2 )
        k4 = h * f ( x + h , y + k3 )
        y += (k1 + 2* k2 + 2* k3 + k4 ) / 6
        x += h
    return y
```
