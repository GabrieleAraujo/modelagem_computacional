-0.800*y^(3/2) + 10*2000*(1 - np.exp(-3*t))

com as seguintes condições iniciais:
y(0) = 2000

Subintervalos: 250, no intervalo de 0 a 2.


````python
#Import package
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
#Derivada
def f(x,y): 
    return -0.800 * y**(3/2) + 10 * 2000 * (1 - np.exp(-3 * x))
````

````python
# #Analítica
# def analitf(t): 
#     return (t+1.0)**2 - np.exp(t)/2.0
````

````python
a = 0
b = 2
h = 2 / 250
y0 = 2000
````

````python
n = int((b - a) / h) + 1
````

# Euler Melhorado

````python
def euler_melhorado(a, b, h, y0, n):
    
    x = np.linspace(a, b, n)
    y = [0] * n
    y[0] = y0

    for i in np.arange(0, n-1):
        k1 = f(x[i], y[i])
        k2 = f(x[i] + h, y[i] + h * k1)
        y[i + 1] = y[i] + (h/2) * (k1 + k2)
                
    return (x,y)
````

````python
x, y = euler_melhorado(a, b, h, y0,n)
print (f'Solução x: {x}')
print (f'Solução y: {y}')
# for point in sol:
#     print(f'x = {point[2]:.4f}, y = {point[3]:.4f}')
````

# Gráfico Euler Melhorado

````python
sns.set(font_scale = 1, font='sans serif')
plt.figure(figsize=(10,6), facecolor = "#FFF")

plt.figure(1)
plt.rcParams.update({'font.size': 10})
plt.plot(x,y, 'b', linewidth = 3)
plt.title('Euler Melhorado')
plt.xlabel('x')
plt.ylabel('y(x)')
#plt.legen(['Euler'])
plt.grid(True)
plt.show()
````

# Ponto Médio

````python
def ponto_medio(a, b, h, y0, n):
    
    x = np.linspace(a, b, n)
    y = [0] * n
    y[0] = y0

    for i in np.arange(0, n-1):
        k1 = f(x[i], y[i])
        k2 = f(x[i] + (h/2), y[i] + k1 * (h/2))
        y[i + 1] = y[i] + k2 * h
                
    return (x,y)
````

````python
x1, y1 = ponto_medio(a, b, h, y0, n)
print (f'Solução x: {x1}')
print (f'Solução y: {y1}')
# for point in sol:
#     print(f'x = {point[2]:.4f}, y = {point[3]:.4f}')
````

````python
sns.set(font_scale = 1, font='sans serif')
plt.figure(figsize=(10,6), facecolor = "#FFF")

plt.figure(1)
plt.rcParams.update({'font.size': 10})
plt.plot(x,y, 'b', linewidth = 3)
plt.title('Ponto Médio')
plt.xlabel('x')
plt.ylabel('y')
#plt.legen(['Euler'])
plt.grid(True)
plt.show()
````

# Comparação 

````python
sns.set(font_scale = 1, font='sans serif')
plt.figure(figsize=(10,6), facecolor = "#FFF")

plt.figure(1)
plt.rcParams.update({'font.size': 10})
plt.plot(x,y, 'b', linewidth = 1, label='Euler Melhorado')
plt.plot(x1,y1, 'r', linewidth = 1, label='Ponto Médio')
plt.xlabel('x')
plt.ylabel('y')
#plt.legen(['Euler'])
plt.legend()
plt.grid(True)
plt.show()
````
