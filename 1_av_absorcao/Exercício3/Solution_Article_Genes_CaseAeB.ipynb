```python
#Parâmetros
w0 = 25
k0 = 0.1425
k1 = 0.4785
k2 = 0.2568
k3 = 0.3691
k4 = 0.7655
t_max = 180
A = (-k0 + k1 + k2 + k3)
delta = A**2 - 4 * (-k0*k2 - k0*k2 + k1*k3)
lambda1 = -k4
lambda2 = (-A + np.sqrt(delta)) / 2
lambda3 = (-A - np.sqrt(delta)) / 2
t = np.linspace(0, t_max, 1000)
```

## DNA

```python
w = (w0 / (lambda3 - lambda2)) * (np.exp(lambda3 * t) * (lambda3 + k2 + k3) - np.exp(lambda2 * t) * (lambda2 + k2 + k3))

plt.plot(t, w)
plt.title('Δ > 0')
plt.xlabel('Time (t)')
plt.ylabel('DNA)')
plt.grid(True)
plt.show()
```

## mRNA

```python
x = (w0 * k1 / (lambda3 - lambda2)) * (np.exp(lambda3 * t) - np.exp(lambda2 * t))

plt.plot(t, x)
plt.title('mRNA (Δ > 0)')
plt.xlabel('Time (t)')
plt.ylabel('mRNA')
plt.grid(True)
plt.show()
```

## Proteína

```python
p = ((w0 * k1 * k3) / (lambda2 * lambda3 * (lambda3 - lambda1))) * np.exp(lambda1 * t) * (
        ((np.exp(lambda3 - lambda1) ** t) - 1) / (lambda3 - lambda1) - ((np.exp(lambda2 - lambda1) ** t) - 1) / (lambda2 - lambda1))

# Plotagem do gráfico
plt.plot(t, p)
plt.title('Proteína (Δ > 0)')
plt.xlabel('Tempo (t)')
plt.ylabel(' Proteína')
plt.grid(True)
plt.show()
```
