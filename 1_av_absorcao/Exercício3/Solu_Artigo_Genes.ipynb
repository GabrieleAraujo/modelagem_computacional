```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
w0 = 1.0  # Initial DNA concentration
k0 = 0.1
k1 = 0.2
k2 = 0.3
k3 = 0.4
k4 = 0.5

# Define the time points for which you want to calculate the solutions
t_max = 180  # Adjust the time range as needed
dt = 0.1  # Time step

# Calculate constants
A = (-k0 + k1 + k2 + k3)
delta = A**2 - 4 * (-k0 * k2 - k0 * k2 + k1 * k3)
lamda1 = -k4
lambda2 = (-A + np.sqrt(delta)) / 2
lambda3 = (-A - np.sqrt(delta)) / 2

# Initialize arrays to store the solutions
num_points = int(t_max / dt) + 1
t = np.linspace(0, t_max, num_points)
w = np.zeros(num_points)
x = np.zeros(num_points)
p = np.zeros(num_points)

# Set initial conditions
w[0] = w0
x[0] = 0
p[0] = 0

# Numerical solution using Euler's method
for i in range(num_points - 1):
    if delta > 0:
        dw = (w[i] / (lambda3 - lambda2)) * (np.exp(lambda3 * t[i]) * (lambda3 + k2 + k3) - np.exp(lambda2 * t[i]) * (lambda2 + k2 + k3)) * dt
        dx = (w[i] * k1 / (lambda3 - lambda2)) * (np.exp(lambda3 * t[i]) - np.exp(lambda2 * t[i])) * dt
        dp = (w[i] * k1 * k3 / (lambda2 * lambda3 * (lambda3 - lambda2))) * np.exp(lamda1 * t[i]) * ((((np.exp(lambda3 - lamda1) * t[i]) - 1) / (lambda3 - lamda1) - (((np.exp(lambda2 - lamda1) * t[i]) - 1) / (lambda2 - lamda1))) * dt

        w[i + 1] = w[i] + dw
        x[i + 1] = x[i] + dx
        p[i + 1] = p[i] + dp

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(t, w, label='DNA Concentration (w)')
plt.plot(t, x, label='mRNA Concentration (x)')
plt.plot(t, p, label='Protein Concentration (p)')
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.legend()
plt.title('Gene Expression Model using Euler Method')
plt.grid(True)
plt.show()
```

## Euler

```python
def euler(f, x, y0, h):
    
    y = np.zeros((len(x), len(y0)))
    f1 = np.zeros((len(x), len(y0)))
    y[0] = y0

    for i in np.arange(0, len(x)-1):
        f1[i] = f(x[i], y[i])
        y[i+1] = y[i] + f1[i] * h
                
    return y           
```

# Runge-kutta

```python

def range_kutta(f, x, y0, h):
    y = np.zeros((len(x), len(y0)))
    k1 = np.zeros((len(x), len(y0)))
    k2 = np.zeros((len(x), len(y0)))
    k3 = np.zeros((len(x), len(y0)))
    k4 = np.zeros((len(x), len(y0)))
    y[0] = y0
    
    for i in np.arange(0, len(x)-1):
                  
        k1[i] = f(x[i],y[i])
        k2[i] = f(x[i] + (h/2), y[i] + (h/2) * k1[i])
        k3[i] = f(x[i] + h, y[i] + h * k2[i])
        k4[i] = f(x[i] + h, y[i] + h * k3[i])
            
        y[i+1] = y[i] + (h/6) * (k1[i] + 2 * k2[i] + 2 * k3[i] + k4[i])

                
    return y   
```
