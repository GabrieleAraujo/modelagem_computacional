# Laplace Bidimensional (Aula 03/11)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Condições iniciais

```python
# Condições iniciais
Lx = 2*np.pi
Ly = 2*np.pi
T = 0.5
nx = 101  # Número de pontos na direção x
ny = 101
nt = 1000  # Número de passos no tempo
dx = Lx/(nx - 1)  # Delta x
dy = Ly/(ny - 1)  # Delta y
dt = T / nt  # Delta t
alpha = dx/dy


x = np.linspace(0, np.pi, nx)
y = np.linspace(0, np.pi, ny)
u = np.zeros((nx, ny))
u[:, 0] = np.sin(x)
u[:,-1] = np.sin(y)
```

## Laplace Bidimensional

```python
for n in range(1000):
    for i in range(1, nx - 1):
        for j in range(1, ny - 1):
            u[i, j] = 0.25 * (u[i + 1, j] + u[i - 1, j] + u[i, j + 1] + u[i, j - 1])
```

```python
#Matriz resultante
plt.imshow(u, extent=[0, Lx, 0, Ly], origin='lower', aspect='auto', cmap='viridis')
plt.colorbar()
plt.title('Laplace Bidimensional')
plt.xlabel('Direção x')
plt.ylabel('Direção y')
plt.show()
```

## Condições de contorno

```python
Z,J= np.meshgrid(x,y)
plt.contourf(Z,J,u,cmap = 'viridis')
plt.colorbar()
plt.title('Laplace Bidimensional')
plt.xlabel('Direção x')
plt.ylabel('Direção y')
```

## Laplace Bidimensional 2

```python
x = np.linspace(0, np.pi, nx)
y = np.linspace(0, np.pi, ny)
v = np.zeros((nx, ny))
v[:, 0] = np.sin(x)
v[:,-1] = np.sin(y)
```

```python
for n in range(1000):
    v_new = np.copy(v)
    for i in range(1, nx - 1):
        for j in range(1, ny - 1):
            v_new[i, j] = (((v[i + 1, j] + v[i - 1, j]) + (alpha**2)*(v[i, j + 1] + v[i, j - 1]))/(2*(1+alpha**2)))
    v = np.copy(v_new)
```

```python
#Matriz resultante
plt.imshow(v, extent=[0, Lx, 0, Ly], origin='lower', aspect='auto', cmap='viridis')
plt.colorbar()
plt.title('Laplace Bidimensional')
plt.xlabel('Direção x')
plt.ylabel('Direção y')
plt.show()
```

## Condições de contorno

```python
X,Y= np.meshgrid(x,y)
plt.contourf(X,Y,v,cmap = 'viridis')
plt.colorbar()
plt.title('Laplace Bidimensional')
plt.xlabel('Direção x')
plt.ylabel('Direção y')
```
